# Text Representation - Bag Of Words (BOW)

```python
import pandas as pd
```

```python
df = pd.read_csv("Spam.csv")
```

```python
df.Category.value_counts()
```

```python
df["Spam"] = df["Category"].apply(lambda x : 1 if x == "spam" else 0)
```

```python
df.drop("Category",axis = 1)
```

```python
df.shape
```

## Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df["Message"],df["Spam"],test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
type(x_train)
```

##  bag of words (BOW's)representation using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X_train_cv = v.fit_transform(x_train.values)
```

```python
import numpy as np
```

```python
X_train_cv.toarray()[:4][0]
```

```python
X_train_cv.shape
```

```python
v.get_feature_names_out()
```

```python
v.vocabulary_
```

```python
v.get_feature_names_out()[2553]
```

## Train the naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(x_test)
```

```python
X_test_cv.toarray()
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))  # use classification_report when their is bais dataset
```

```python
model.predict(X_test_cv)
```

```python
email =["You've been selected for a free $500 gift card! Act fast and claim it by clicking the link below:"]
emails_count = v.transform(email)
model.predict(emails_count)
```

## Train the model using sklearn pipeline and reduce number of lines of code

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)

print(classification_report(y_test, y_pred))
```
