```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("IMDB Dataset.csv")
```

```python
df
```

```python
df.shape
```

```python
df.sentiment.value_counts()
```

```python
df["sentiment"] = df["sentiment"].apply(lambda x : 1 if x == "negative" else 0)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df["review"],df["sentiment"],test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Using MultinomialNB Classification Algo

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = clf.predict(x_test)

print(classification_report(y_test, y_pred))
```

## Using RandomForestClassifier Classification Algo

'random_forest': The label/name for the model.

RandomForestClassifier: A machine learning model that uses multiple decision trees.

n_estimators=50: Creates 50 decision trees.

criterion='entropy': Uses entropy to measure split quality.

```python
from sklearn.ensemble import RandomForestClassifier

clf = Pipeline([
    ('vectorizer', CountVectorizer()),                                                    
    ('random_forest', (RandomForestClassifier(n_estimators=50, criterion='entropy')))      
])

clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)

print(classification_report(y_test, y_pred))
```
